# Import Library
- cv2: OpenCV untuk pengolahan gambar.
- numpy: Untuk operasi array.
- matplotlib.pyplot: Untuk visualisasi gambar.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Membaca, Konversi dan Thresholding

### Membaca Gambar:
- cv2.imread() membaca gambar dari file '/content/sample1.jpg'.

### Konversi ke Grayscale:
- cv2.cvtColor() mengubah gambar dari format BGR (default OpenCV) ke grayscale untuk mempermudah pengolahan.

### Thresholding:
- cv2.threshold() mengubah gambar grayscale menjadi citra biner (hitam-putih) dengan nilai ambang 127. Pixel dengan nilai di atas 127 diubah menjadi 255 (putih), dan di bawah menjadi 0 (hitam).

```python
# Membaca gambar dari file
image = cv2.imread('/content/sample1.jpg')

# Mengubah gambar ke grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding untuk mengubah ke citra biner
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

# Deteksi dan Gambar Kontur

### Deteksi Kontur:
- cv2.findContours() mendeteksi kontur (garis tepi objek) dari citra biner. Parameter RETR_EXTERNAL hanya mengambil kontur terluar, dan CHAIN_APPROX_SIMPLE menyederhanakan penyimpanan titik kontur.

### Menggambar Kontur:
- cv2.drawContours() menggambar kontur yang terdeteksi pada salinan gambar asli dengan warna hijau dan ketebalan 2 pixel.

```python
# Deteksi kontur
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Gambar kontur pada gambar asli
image_contour = image.copy()
cv2.drawContours(image_contour, contours, -1, (0, 255, 0), 2)
```

# Analisis Kontur

### Loop melalui setiap kontur yang terdeteksi:

- cv2.contourArea() menghitung luas area kontur.

- cv2.arcLength() menghitung keliling kontur.

- cv2.boundingRect() mendapatkan bounding box (kotak pembatas) kontur, lalu rasio aspek dihitung dari lebar (w) dan tinggi (h) bounding box.

- Hasil luas, keliling, dan rasio aspek dicetak untuk setiap kontur.

```python
for i, cnt in enumerate(contours):
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt, True)

    # Mendapatkan bounding box untuk menghitung rasio aspek
    x, y, w, h = cv2.boundingRect(cnt)
    aspect_ratio = float(w) / h if h != 0 else 0

    print(f"Kontur {i+1}:")
    print(f" - Luas: {area:.2f}")
    print(f" - Keliling: {perimeter:.2f}")
    print(f" - Rasio Aspek: {aspect_ratio:.2f}")
    print("---")
```
