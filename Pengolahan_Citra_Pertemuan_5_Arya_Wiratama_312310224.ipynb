# 3. Mengubah size panjang dan lebar gambar
- filename adalah nama dari file gambar yang ada
- `with Image.open(filename) as image` adalah langkah untuk membuka gambar

# 5. Memutar gambar
- `img = img.rotate(180)` di dalam PIL terdapat sebuah method untuk memutar gambar `rotate()`, yang di dalamnnya di isi dengan derajat rotatenya `rotate(180)`

```python
from PIL import Image

def main():
    try:
        #Relative Path
        img = Image.open("view.jpg")

        #Angle given
        img = img.rotate(180)

         #Saved in the same relative location
        img.save("rotated_picture.jpg")
    except IOError:
        pass

if __name__ == "__main__":
    main()
```

# 6. Memotong gambar
- `img = img.crop(area)` di dalam Library PIL terdapat sebuah method untuk memotong gambar `crop(area)`, lalu area adalah besar daerah yang akan di potong, pada case kali ini `area` adalah `area = (0, 0, width/2, height/2)`

```python
from PIL import Image

def main():
    try:
        #Relative Path
        img = Image.open("view.jpg")
        width, height = img.size

        area = (0, 0, width/2, height/2)
        img = img.crop(area)

        #Saved in the same relative location
        img.save("cropped_picture.jpg")

    except IOError:
        pass

if __name__ == "__main__":
    main()
```

# 7. Mengubah ukuran gambar
- `img = img.resize((width/2, height/2))` terdapat sebuah method di library PIL untuk mengubah ukuran `resize(untuk_lebar, untuk_tinggi)`, pada case kali ini saya lebar dan tinggi nya adalah `(width//2, height//2)`, yaitu ukuran asli di bagi dengan 2

```python
from PIL import Image

def main():
    try:
         #Relative Path
        img = Image.open("view.jpg")
        width, height = img.size

        img = img.resize((width//2, height//2))

        #Saved in the same relative location
        img.save("resized_picture.jpg")
    except IOError:
        pass

if __name__ == "__main__":
    main()
```

# 8. Menempelkan satu gambar di atas gambar lain (pasting)
- `img.paste()` menempelkan img2 ke img pada posisi (x, y) = (50, 50) (koordinat piksel dari sudut kiri atas).

```python
from PIL import Image

def main():
    try:
        #Relative Path
        #Image on which we want to paste
        img = Image.open("view.jpg")

        #Relative Path
        #Image which we want to paste
        img2 = Image.open("view2.jpg")
        img.paste(img2, (50, 50))

        #Saved in the same relative location
        img.save("pastedd_picture.jpg")

    except IOError:
        pass

if __name__ == "__main__":
    main()

##An additional argument for an optional image mask image is also available.
```

# 9. Membuka sebuah gambar dan mencetak histogram dari gambar
- `img.histogram()` menghitung distribusi intensitas piksel dalam gambar.

```python
def main():
    try:
        #Relative Path
        img = Image.open("view.jpg")

        #Getting histogram of image
        print(img.histogram())

    except IOError:
        pass

if __name__ == "__main__":
    main()
```

# 10. Melakukan operasi transpose (pencerminan) pada gambar
- `img.transpose()` melakukan operasi transformasi geometrik pada gambar.
- `Image.FLIP_LEFT_RIGHT`: Argumen untuk mencerminkan gambar kiri-kanan (horizontal).

```python
from PIL import Image

def main():
    try:
        #Relative Path
        img = Image.open("view.jpg")

        #transposing image
        transposed_img = img.transpose(Image.FLIP_LEFT_RIGHT)

        #Save transposed image
        transposed_img.save("transposed.jpg")
    except IOError:
        pass

if __name__ == "__main__":
    main()
```

# 11. Melakukan operasi split
- `img.split()` memisahkan gambar menjadi channel-channel warna individual.

```python
from PIL import Image

def main():
    try:
        #Relative Path
        img = Image.open("view.jpg")

        #splitting the image

        print(img.split())
    except IOError:
        pass

if __name__ == "__main__":
    main()
```

# 12. Mengonversi ke bitmap
- `img.mode` mengembalikan string yang menunjukkan mode warna gambar.
- `print(img.tobitmap())` mengonversi gambar ke format bitmap (hanya untuk mode '1' atau grayscale 1-bit).

```python
from PIL import Image

def main():
    try:
        img = Image.open("view.jpg")
        img = img.convert('1')  # Konversi ke 1-bit
        print(img.tobitmap())  # Sekarang berhasil
    except IOError:
        print("File tidak ditemukan")

if __name__ == "__main__":
    main()
```

# 13. Membuat thumbnail
- `img.thumbnail((max_width, max_height))` mengubah ukuran gambar secara langsung (in-place) menjadi versi kecil (thumbnail)

```python
def main():
    try:
        #Relative Path
        img = Image.open("view.jpg")

        #In-place modification
        img.thumbnail((200, 200))

        img.save("thumb.jpg")
    except IOError:
        pass

if __name__ == "__main__":
    main()
```
